# Step 1: Load Data

```python
import pandas as pd
import numpy as np 
```

```python
data = pd.read_csv('OnlineRetail.csv',encoding = 'ISO-8859-1')
demo = data.copy()
demo.head()
```

# Step 2: Drop Missing Invoices or Descriptions

```python
demo[['InvoiceDate','Description']].isnull().sum()
```

```python
demo[['InvoiceDate','Description']].dropna()
```

# Step 3: Remove Canceled Invoices <BR> Start with letter 'c'

```python
RemoveC = demo[(demo["InvoiceNo"].astype(str).str.startswith("c"))].dropna()
print(RemoveC)
```

# Step 4: Keep Only Positive Quantities

```python
OnlyPositive = demo[demo["Quantity"]>0]
print(OnlyPositive)
```

# Step 5: Standardize Item Descriptions | Lower case and remove space

```python
demo["Description"]= demo["Description"].astype(str).str.lower().str.strip()
print(demo)
```

# Step 6: Filter for a Single Country (e.g., United Kingdom)

```python
UK = demo[demo['Country'] == 'United Kingdom'] ##â†’ uses a boolean mask to filter rows.
```

```python
demo.query("Country == 'United Kingdom'")
```

# Step 7: Remove Duplicates

```python
demo.duplicated()
```

```python
demo.drop_duplicates()
```

# Step 8: Add TotalPrice column , TotalPrice = Quantity *  UnitPrice

```python
demo['TotalPrice'] = demo['Quantity'] * demo['UnitPrice']
demo
```

# Step 9 : Check for Top 10 Selling Products

```python
topProducts = demo.groupby('Description')['Quantity'].sum().sort_values(ascending=False).head(10)

print(topProducts)
```

# Step 10 : Identify Most Active Customers

```python
demo["CustomerID"].value_counts().head(10)  ##-> Overcounts customers if they bought multiple items in a single invoice.
```

```python
mostactive= demo.groupby('CustomerID')['InvoiceNo'].nunique().sort_values(ascending=False).head(10)  ##-> 
mostactive
```

# Step 11: Time-Based Analysis | Convert InvoiceDate to datetime and check popular months/hours.

```python
demo["InvoiceDate"] = pd.to_datetime(demo["InvoiceDate"])
demo["Months"] = demo['InvoiceDate'].dt.month
demo["Hour"] = demo['InvoiceDate'].dt.hour

print("Popular month",end="\n")
print(demo["Months"].value_counts().head(1))

print("Popular hour",end="\n")
print(demo["Hour"].value_counts().head(1))
```
